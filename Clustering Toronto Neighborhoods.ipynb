### Cluster Borough

```python
!conda install -c conda-forge folium=0.5.0 --yes
import folium # map rendering library
```

```python
!conda install -c conda-forge geopy --yes 
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="Toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Manhattan are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(cdfg['Latitude'], cdfg['Longitude'], cdfg['Borough'], cdfg['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

### Cluster only boroughs that contain the word Toronto 

```python
toronto_grouped= cdfg
toronto_grouped
```

```python
for i in range(toronto_grouped.shape[0]):
    if 'Toronto' not in toronto_grouped['Neighbourhood'][i]:
        toronto_grouped.drop(index=i, inplace=True)
        
toronto_grouped
```

```python
toronto_grouped_c = toronto_grouped.groupby('Borough', as_index=False).mean()
toronto_grouped_c
```

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped_c.drop('Borough', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:5] 
```

```python
# add clustering labels
toronto_grouped_c.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
toronto_grouped_c
```

### Map out Clustered bouroughs 

```python
# create map
map_clusters1 = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_grouped_c['Latitude'], toronto_grouped_c['Longitude'], toronto_grouped_c['Borough'], toronto_grouped_c['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters1)
    
    
       
map_clusters1
```
